## Data Processing

If you are interested in another local authority, then change the values of the variable below:

```python
relevant_region = 'Greenwich'
comparison_regions = ['Bromley', 'Bexley', 'Lewisham', 'Tower Hamlets'] # this is in the order you would like to appear in bar charts & legends
broader_region = 'London'
country = 'England' # changing this won't work, as countries have different sheets
```

```python
def add_rating_percentages(df):
  df["A%"] = df["A"] / df["Number Lodgements"] * 100
  df["B%"] = df["B"] / df["Number Lodgements"] * 100
  df["C%"] = df["C"] / df["Number Lodgements"] * 100
  df["D%"] = df["D"] / df["Number Lodgements"] * 100
  df["E%"] = df["E"] / df["Number Lodgements"] * 100
  df["F%"] = df["F"] / df["Number Lodgements"] * 100
  df["G%"] = df["G"] / df["Number Lodgements"] * 100
  df["Not Recorded%"] = df["Not Recorded"] / df["Number Lodgements"] * 100
    
def add_co2_numbers(df):
  df["CO2 Emissions per home"] = df["Co 2 Emissions Current"] / df["Number Lodgements"]
  df["CO2 Emissions per m2"] = df["Co 2 Emissions Current"] / df["Total Floor Area"] * 1000
  df["Floor Area per home"] = df["Total Floor Area"] / df["Number Lodgements"]
  df["Floor Area per home (sq ft)"] = df["Total Floor Area"] / df["Number Lodgements"] * 10.7639
    
def filter_region(df):
  return df.loc[df['Region'] == relevant_region]

def add_year(df):
  df["Year"] = df["Quarter"].str[:4]
    
def group_by_year(df, cols):
  return df[cols].groupby(['Year']).sum().reset_index()

def group_ratings_by_year(df):
  by_year_df = group_by_year(df, ["Year", "Number Lodgements", "Total Floor Area", "A", "B", "C", "D", "E", "F", "G", "Not Recorded"])
  add_rating_percentages(by_year_df)
  return by_year_df

def group_co2_by_year(df):
  by_year_df = group_by_year(df, ["Year", "Number Lodgements", "Total Floor Area", "Co 2 Emissions Current"])
  add_co2_numbers(by_year_df)
  return by_year_df

def group_regional_comparison_co2_by_year(df):
  by_year_df = df[["Region", "Year", "Number Lodgements", "Total Floor Area", "Co 2 Emissions Current"]].groupby(['Region', 'Year']).sum().reset_index()
  add_co2_numbers(by_year_df)
  return by_year_df

def combine_new_and_existing(new_df, existing_df):
  new_df['Type'] = 'New'
  existing_df['Type'] = 'Existing'
  return pd.concat([new_df, existing_df])

def filter_comparison_regions(df):
  return df.loc[df['Region'].isin([relevant_region] + comparison_regions)]

def filter_broader_region(df):
  df_1 = df.loc[df['Region'] == broader_region]
  return df_1.loc[df_1['Quarter'] != None]

def filter_country(df):
  return df.loc[df['Quarter'] != None]

def combine_broader_regions(region_df, broader_region_df, country_df):
  region_df['Region'] = relevant_region
  broader_region_df['Region'] = broader_region
  country_df['Region'] = country
  return pd.concat([region_df, broader_region_df, country_df])

add_rating_percentages(ratings_df)
add_year(ratings_df)
regional_ratings_df = filter_region(ratings_df)
regional_ratings_by_year_df = group_ratings_by_year(regional_ratings_df)

add_rating_percentages(new_ratings_df)
add_year(new_ratings_df)
regional_new_ratings_df = filter_region(new_ratings_df)
regional_new_ratings_by_year_df = group_ratings_by_year(regional_new_ratings_df)

add_rating_percentages(existing_ratings_df)
add_year(existing_ratings_df)
regional_existing_ratings_df = filter_region(existing_ratings_df)
regional_existing_ratings_by_year_df = group_ratings_by_year(regional_existing_ratings_df)

regional_new_existing_ratings_by_year_df = combine_new_and_existing(regional_new_ratings_by_year_df, regional_existing_ratings_by_year_df)

add_co2_numbers(co2_df)
add_year(co2_df)
regional_co2_df = filter_region(co2_df)
regional_co2_by_year_df = group_co2_by_year(regional_co2_df)

add_co2_numbers(new_co2_df)
add_year(new_co2_df)
regional_new_co2_df = filter_region(new_co2_df)
regional_new_co2_by_year_df = group_co2_by_year(regional_new_co2_df)

add_co2_numbers(existing_co2_df)
add_year(existing_co2_df)
regional_existing_co2_df = filter_region(existing_co2_df)
regional_existing_co2_by_year_df = group_co2_by_year(regional_existing_co2_df)

regional_new_existing_co2_by_year_df = combine_new_and_existing(regional_new_co2_by_year_df, regional_existing_co2_by_year_df)

regional_comparison_new_co2_df = filter_comparison_regions(new_co2_df)
regional_comparison_new_co2_by_year_df = group_regional_comparison_co2_by_year(regional_comparison_new_co2_df)

regional_comparison_existing_co2_df = filter_comparison_regions(existing_co2_df)
regional_comparison_existing_co2_by_year_df = group_regional_comparison_co2_by_year(regional_comparison_existing_co2_df)

add_co2_numbers(new_co2_region_df)
add_year(new_co2_region_df)
broader_region_new_co2_df = filter_broader_region(new_co2_region_df)
broader_region_new_co2_by_year_df = group_co2_by_year(broader_region_new_co2_df)

add_co2_numbers(new_co2_country_df)
add_year(new_co2_country_df)
country_new_co2_df = filter_country(new_co2_country_df)
country_new_co2_by_year_df = group_co2_by_year(country_new_co2_df)

broader_comparison_new_co2_by_year_df = combine_broader_regions(regional_new_co2_by_year_df, broader_region_new_co2_by_year_df, country_new_co2_by_year_df)

add_co2_numbers(existing_co2_region_df)
add_year(existing_co2_region_df)
broader_region_existing_co2_df = filter_broader_region(existing_co2_region_df)
broader_region_existing_co2_by_year_df = group_co2_by_year(broader_region_existing_co2_df)

add_co2_numbers(existing_co2_country_df)
add_year(existing_co2_country_df)
country_existing_co2_df = filter_country(existing_co2_country_df)
country_existing_co2_by_year_df = group_co2_by_year(country_existing_co2_df)

broader_comparison_existing_co2_by_year_df = combine_broader_regions(regional_existing_co2_by_year_df, broader_region_existing_co2_by_year_df, country_existing_co2_by_year_df)
```

## A high EPC doesn't mean low emissions

*The below is based on my understanding, and would need to be checked with someone who knows what they're talking about if we were to ever share publicly.*

The EPC rating is really a rating of how much it will cost to run your home to a specified standard of living per m2 of living space. This is related to CO2, but isn't the same.

For example, using the energy values from [MoneySavingExpert](https://www.moneysavingexpert.com/utilities/what-is-the-energy-price-cap/) and the government's [2023 conversion factors](https://www.gov.uk/government/publications/greenhouse-gas-reporting-conversion-factors-2023), if you required 10,000 kWh of energy to heat your home every year, then:
    
| Heating Method     | Energy Required / kWh | Cost per kWh / p | Total cost per year / Â£ | Emissions per kWh / kg CO2 equivalent | Total emissions per year / tonnes CO2 equivalent |
|--------------------|-----------------------|------------------|-------------------------|---------------------------------------|--------------------------------------------------|
| Gas boiler         | 10,000 kWh            | 6.89p            | Â£689                    | 0.203 kg                              | 2.03 t                                           |
| Electric radiators | 10,000 kWh            | 27.35p           | Â£2,735                  | 0.207 kg                              | 2.07 t                                           |
| Electric heat pump | 3,000 kWh             | 27.35p           | Â£821                    | 0.207 kg                              | 0.62 t                                           |

This would mean that electric radiators would be given a far worse EPC rating, and the heat pump would be given a slightly worse EPC rating than the gas boiler; despite electric radiators emitting a similar amount of carbon, and the heat pumps emitting significantly less carbon than the gas boiler.

This discrepancy will only become more pronounced as the grid decarbonises, because the emissions per kWh of electricity is continuously falling.
